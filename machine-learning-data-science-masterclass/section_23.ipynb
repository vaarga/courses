# Section 23: Hierarchical Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy
```

## Data and Visualization

```python
df = pd.read_csv('cluster_mpg.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_w_dummies = pd.get_dummies(df.drop('name', axis=1))
```

```python
df_w_dummies
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_w_dummies)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=df_w_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(15,8))

sns.heatmap(scaled_df)
```

```python
plt.figure(figsize=(15,8))

sns.clustermap(scaled_df, col_cluster=False)
```

```python
sns.heatmap(scaled_df.corr())
```

## Clustering

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(12,6))

sns.scatterplot(df, x='mpg', y='weight', hue=cluster_labels)
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold=2)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(14,8))

dendrogram = hierarchy.dendrogram(linkage_matrix, truncate_mode='lastp', p=11)
```
