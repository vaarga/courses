## Styling

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')
ax.plot(x, x ** 2, label='X vs X^2')
ax.legend(loc=(1.1,0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color='#0362fc', linewidth=5, marker='+', ms=10, ls='--')
ax.plot(x, x + 1, color='#ffc600', linewidth=3, marker='o', markersize=10, linestyle='-.',
        markerfacecolor='red', markeredgewidth=4, markeredgecolor='blue')
```

## Exercises

$$ E = mc^2 $$

```python
c = 300000000
m = np.linspace(0, 10, 11)
m
```

```python
E = m * (c ** 2)
E
```

```python
fig = plt.figure(figsize=(0.8, 0.5), dpi=100)

ax = fig.add_axes([0, 0, 8, 8])

ax.plot(m, E, color='red', lw=5)
ax.set_xlim(0, 10)
ax.set_xlabel('Mass in Grams')
ax.set_ylabel('Energy in joules')
ax.set_title('E=mc^2')
```

```python
plt.plot(m, E, color='red', lw=10)
plt.xlim(0, 10)
plt.grid(True, which='both', axis='y')
plt.yscale('log')
plt.xlabel('Mass in Grams')
plt.ylabel('Energy in joules')
plt.title('E=mc^2')
```

```python
labels = ['1 Mo', '3 Mo', '6 Mo', '1 Yr', '2 Yr', '3 Yr', '5 Yr', '7 Yr', '10 Yr', '20 Yr', '30 Yr']
july16_2007 =[4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.plot(labels, july16_2007, label='july16_2007')
plt.plot(labels, july16_2020, label='july16_2020')
plt.legend()
plt.show()
```

```python
plt.plot(labels, july16_2007, label='july16_2007')
plt.plot(labels, july16_2020, label='july16_2020')
plt.legend(loc=(1.1, 0.5))
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(labels, july16_2007)
ax[0].set_title('July 16th, 2007')
ax[1].plot(labels, july16_2020)
ax[1].set_title('July 16th, 2020')

fig.tight_layout()
```

```python
fig, ax1 = plt.subplots()

ax1.plot(labels, july16_2007, color='blue')
ax1.set_title('July 16th Yield Curves')
ax1.set_ylabel('2007', color='blue')

for label in ax1.get_yticklabels():
        label.set_color('blue')

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, color='red')
ax2.set_ylabel('2020', color='red')

for label in ax2.get_yticklabels():
        label.set_color('red')
        
ax2.spines['left'].set_color('blue')
ax2.spines['left'].set_linewidth(2)
ax2.spines['right'].set_color('red')
ax2.spines['right'].set_linewidth(2)
```
