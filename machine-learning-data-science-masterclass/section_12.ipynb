## Grid Search

```python
df = pd.read_csv('Advertising.csv')
```

```python
X = df.drop('sales', axis=1)
```

```python
y = df['sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
base_elastic_net_model = ElasticNet()
```

```python
param_grid = {
    'alpha': [.1, 1, 5, 10, 50, 100],
    'l1_ratio': [.1, .5, .7, .95, .99, 1],
}
```

```python
grid_model = GridSearchCV(base_elastic_net_model, param_grid, scoring='neg_mean_squared_error', cv=5, verbose=1)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_pred = grid_model.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```

## Linear Regression Project

```python
df = pd.read_csv('Ames_Housing_Data_Final.csv', index_col=0)
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop('SalePrice', axis=1)
```

```python
y = df['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
model = ElasticNet()
```

```python
param_grid = {
    'alpha': [.1, 1, 5, 10, 100],
    'l1_ratio': [.1, .7, .99, 1],
}
```

```python
grid_model = GridSearchCV(model, param_grid, scoring='neg_mean_squared_error', cv=5, verbose=2)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
final_model = grid_model.best_estimator_
```

```python
y_pred = final_model.predict(X_test)
```

```python
mean_absolute_error(y_pred, y_test)
```

```python
root_mean_squared_error(y_pred, y_test)
```
