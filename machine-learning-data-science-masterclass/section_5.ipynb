## Apply

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
def get_last_four_digits(number):
    return str(number)[-4:]
```

```python
get_last_four_digits(123456789)
```

```python
df['last_four_digits'] = df['CC Number'].apply(get_last_four_digits)
```

```python
df
```

```python
df['total_bill'].mean()
```

```python
def convert_price(price):
    if price < 10:
        return '$'
    elif price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['converted_price'] = df['total_bill'].apply(convert_price)
```

```python
df
```

```python
df['total_bill'].apply(lambda number: number * 2)
```

```python
def get_tip_quality(total_bill, tip):
    if tip / total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
```

```python
df['tip_quality'] = df[['total_bill', 'tip']].apply(lambda local_df: get_tip_quality(local_df['total_bill'], local_df['tip']), axis=1)
```

```python
df['tip_quality'] = np.vectorize(get_tip_quality)(df['total_bill'], df['tip'])
```

```python
df.head()
```
