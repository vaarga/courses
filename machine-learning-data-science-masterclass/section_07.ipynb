```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
plt.figure(figsize=(12, 4))

sns.boxplot(df, y='parental level of education', x='reading score', hue='test preparation course', palette='Set2')

plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
sns.violinplot(df, x='reading score', y='parental level of education', bw_adjust=0.5)
```

```python
plt.figure(figsize=(8, 4), dpi=200)

sns.swarmplot(df, x='math score', y='gender', size=2, hue='test preparation course', dodge=True)    
```

```python
sns.boxenplot(df, x='math score', y='test preparation course', hue='gender')
```

```python
sns.jointplot(df, x='math score', y='reading score', hue='gender')
```

```python
sns.pairplot(df, hue='gender', corner=True)
```

```python
sns.catplot(df, x='gender', y='math score', kind='violin', col='lunch', row='test preparation course')
```

```python
g = sns.PairGrid(df, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend()
```

```python
df = pd.read_csv('application_record.csv')
df
```

```python
df.info()
```

```python
df['DAYS_BIRTH'] = df['DAYS_BIRTH'] * -1
df['DAYS_EMPLOYED'] = df['DAYS_EMPLOYED'] * -1
```

```python
plt.figure(figsize=(12, 6), dpi=200)

sns.scatterplot(df[df['DAYS_EMPLOYED'] > 0], x='DAYS_BIRTH', y='DAYS_EMPLOYED', alpha=0.01, lw=0)
```

```python
df['YEARS_BIRTH'] = df['DAYS_BIRTH'] / 365
df['YEARS_BIRTH']
```

```python
sns.displot(df, x='YEARS_BIRTH', bins=45, aspect=10/4)
```

```python
df_len = len(df)
df_len
```

```python
bottom_half_income = df.nsmallest(int(df_len/2), 'AMT_INCOME_TOTAL')
```

```python
plt.figure(figsize=(10, 5))

sns.boxplot(bottom_half_income, x='NAME_FAMILY_STATUS', y='AMT_INCOME_TOTAL', hue='FLAG_OWN_REALTY')

plt.legend(loc=(1.05, 0.5), title='FLAG_OWN_REALTY')
```
